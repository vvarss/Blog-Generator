```python
import os 
import openai
from openai import OpenAI
```

```python
os.environ["OPENAI_API_KEY"] = "YOUR_API_KEY"
client = OpenAI()
```

```python
print(client.models.list())
```

```python
client.completions.create(
    model="gpt-4o-mini",
    prompt="Tell me about the sky",
)
```

```python
#Building the blog generator
```

```python
response = client.completions.create(
    model="gpt-4o-mini",
    prompt="Generate a blog about life",
    temperature=0,                 #Gives a precise response , if kept 1, it will be creative, gives response in bullets.
    max_tokens=700,
)
```

```python
print(response)
```

```python
print(response.choices[0].text)
```

```python
# It has stopped here, because we have assigned the max_tokens as 700
```

```python
#We can also prompt a persona, setting the context and generate the content.
```

```python
prompt = """
You are a software developer with 4 years of coding experience.
Your task is to share information about your most favourite subject during your academic years.

Topic:Operating Systems
additional points:None
"""
response = client.completions.create(
    model="gpt-4o-mini",
    prompt=prompt,
    temperature=1,
    max_tokens=200,
)
```

```python
print(response.choices[0].text)
```

```python
# Accepting User input
```

```python
topic = input("Enter the topic for the blog:")
additional_pointer = input("Enter any additional points, if any:")
prompt = f"""
Your task is to write a blog, with subtopics accordingly.
Each paragraph should be two sentences long.
topic:{topic}
additional_pointer:{additional_pointer}
"""
response = client.completions.create(
    model="gpt-4o-mini",
    prompt=prompt,
    temperature=0,
    max_tokens=400,
)
print(response.choices[0].text)
```
